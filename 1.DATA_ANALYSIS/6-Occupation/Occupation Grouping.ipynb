# Occupation

```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users 
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")[["age"]].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# NÃºmero de hombres en cada empleo
n_hombres = users.loc[users.loc[:,"gender"] == "M"].groupby("occupation")[["gender"]].value_counts()
n_hombres
```

```python
# NÃºmero total de empleos
n_empleos = users.loc[:,"occupation"].value_counts()
n_empleos
```

```python
pd.DataFrame(n_hombres/n_empleos).reset_index().sort_values(by = "count", ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")[["age"]].agg(["min","max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"])[["age"]].mean().reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
pd.DataFrame(users.loc[(users.loc[:,"gender"] == "M") | (users.loc[:,"gender"] == "F")].groupby("occupation")[["gender"]].value_counts()/n_empleos).reset_index()
```
