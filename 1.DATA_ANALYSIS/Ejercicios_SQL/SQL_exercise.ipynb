```python
import pandas as pd
import sqlite3
```

```python
# Conectamos con la base de datos
connection = sqlite3.connect("data/chinook.db")

# Obtenemos un cursor que utilizaremos para hacer las queries
crsr = connection.cursor()
```

```python
# Función para leer los datos y lo pasarlos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

```python
# 7. Muestra cuantas canciones hay en cada playlist y el nombre de cada playlist
querty = """ 
SELECT playlists.Name, COUNT(tracks.Name) AS Num_Songs

FROM tracks
INNER JOIN playlist_track
    ON tracks.TrackId = playlist_track.TrackId
INNER JOIN playlists
    ON playlist_track.PlaylistId = playlists.PlaylistId

GROUP BY playlists.Name
"""

sql_query(querty)
```

```python
# 8. Muestra cuánto ha vendido cada empleado
query = """
SELECT DISTINCT employees.FirstName, employees.LastName, SUM(Quantity) AS Total_Sales

FROM employees
INNER JOIN customers
    ON employees.EmployeeId = customers.SupportRepid
INNER JOIN invoices
    ON customers.CustomerId = invoices.CustomerId
INNER JOIN invoice_items
    ON invoices.InvoiceId = invoice_items.InvoiceId

GROUP BY employees.FirstName, employees.LastName

"""

sql_query(query)
```

```python
# 9. Quien ha sido el agente de ventas que más ha vendido en 2009?
query = """
SELECT DISTINCT employees.FirstName, employees.LastName, SUM(Quantity) AS Total_Sales

FROM employees
INNER JOIN customers
    ON employees.EmployeeId = customers.SupportRepid
INNER JOIN invoices
    ON customers.CustomerId = invoices.CustomerId
INNER JOIN invoice_items
    ON invoices.InvoiceId = invoice_items.InvoiceId

GROUP BY employees.FirstName, employees.LastName
ORDER BY Total_Sales DESC

LIMIT 1
"""

sql_query(query)
```

```python
# 10. Quien es son los 3 grupos que más han vendido?
query = """ 
SELECT artists.Name,SUM(invoice_items.UnitPrice * invoice_items.Quantity) AS Total_Sales 

FROM artists
INNER JOIN albums
    ON artists.ArtistId = albums.ArtistId
INNER JOIN tracks
    ON albums.AlbumId = tracks.AlbumId
INNER JOIN invoice_items
    ON tracks.TrackId = invoice_items.TrackId

GROUP BY artists.Name
ORDER BY Total_Sales DESC 

LIMIT 3
"""

sql_query(query)
```
