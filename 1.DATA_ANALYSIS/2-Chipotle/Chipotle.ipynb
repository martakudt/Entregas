### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

Fijaos que es un tsv. Eso significa "Tab separated value". IMPORTANTE: El separador es "\t"

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep="\t")
```

```python
chipo
```

### Step Extra:

Si os fijais, la columna item_price es un string. Hay que quitar el signo del dolar y convertir a float para poder trabajar. 

Hay maneras cool de hacer esto, pero os recomiendo lo siguiente:
- haced un bucle for del df (iterrows)
- guardaros en una lista (append) el valor sin el dolar y convetido a float
- actualizad el valor de esa columna con la nueva lista

```python
new_item_price = []
for index,row in chipo.iterrows():
    new_price = row.loc["item_price"].replace("$", "")
    new_item_price.append(float(new_price))
new_item_price
```

```python
def clean_price(price):
    return float(price[1:])

chipo.loc[:"item_price"].apply(clean_price)
```

```python
chipo.loc[:,"item_price"] = new_item_price
chipo
```

### Step 3. Name of the max valued product

```python
# Faltaba sacar el precio por unidad
chipo.loc[:, "unity_price"] = chipo.loc[:,"item_price"]/chipo.loc[:,"quantity"]
chipo
```

```python
maximo = chipo.loc[:,"unity_price"].max()
chipo.loc[chipo.loc[:,"unity_price"] == maximo]["item_name"].iloc[0]
```

### Step 4. How many products cost more than $10.00?

```python
productos_10 = chipo.loc[chipo["unity_price"]>10.00]
len(productos_10)
```

### Step 5. a - How much money we earned selling our most expensive product?

Busca el producto mas caro (si hay mas de uno da igual, coge el uno de ellos). Guardate su nombre y precio.
Filtra las filas por ese producto con el nombre.
Multiplica las cantidades del resultado filtrado por el precio que guardaste.


```python
nombre = chipo.loc[chipo.loc[:,"unity_price"] == maximo]["item_name"].iloc[0]
precio = chipo.loc[chipo.loc[:,"unity_price"] == maximo]["unity_price"].iloc[0]

filtrado = chipo.loc[chipo.loc[:,"item_name"] == nombre]
filtrado.loc[:,"quantity"].sum()*precio
```

### Step 5. b - How much money we earned selling our cheapest product?

Lo mismo que antes pero con el mas barato.

```python
minimo = chipo.loc[:,"unity_price"].min()

nombre2 = chipo.loc[chipo.loc[:,"unity_price"] == minimo]["item_name"].iloc[0]
precio = chipo.loc[chipo.loc[:,"unity_price"] == minimo]["unity_price"].iloc[0]

filtrado = chipo.loc[chipo["item_name"] == nombre2]
filtrado.loc[:,"quantity"].sum()*precio
```
