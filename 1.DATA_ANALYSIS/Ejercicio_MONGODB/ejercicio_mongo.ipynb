```python
url = "mongodb+srv://admin:admin@cluster0.jpwo6w9.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"
myclient = pymongo.MongoClient(url)
mydb = myclient["Cluster0"] # INSTANCIAR BASE DE DATOS
restaurantes = mydb['restaurantes']
```

```python
# 1. Mostrar todos los documentos de la colección restaurantes
# filtro = {"cuisine": "Bakery", "borough": "Bronx"}
filtro = {}
proyeccion = {"name": 1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 2. Mostrar los campos restaurant_id, nombre, distrito y cocina, pero excluya el campo _id para todos los documentos de la colección restaurantes
list(restaurantes.find({} , {'_id':0,'restaurant_id':1,'name':1,'borough':1,'cuisine':1}))
```

```python
# 3. Mostrar los primeros 5 restaurantes que se encuentran en el distrito Bronx
list(restaurantes.find({'borough':'Bronx'},{'_id':0}).limit(5))
```

```python
list(restaurantes.find({'borough':'Bronx'},{'_id':0}).limit(5).sort("name"))
```

```python
# 4. Devolver los restaurantes que lograron una puntuación superior a 80 pero inferior a 100
filtro = {"grades": {"$elemMatch": {"score": {"$gt": 80, "$lt": 100}}}}

proyeccion = {"name":1, "grades": 1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 5. Devolver los restaurantes que se ubican en un valor de latitud inferior a -95.754168
filtro = {"address.coord.0": {"$lt": -95.754168}}

proyeccion = {"name":1, "coord": 1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 6. Devolver los restaurantes que no preparan cocina americana y lograron una puntuación superior a 70 y se ubicaron en una longitud inferior a -65.754168. Nota: Realice esta consulta sin usar el operador $and
filtro = {
    "cuisine": {"$not": {"$regex": "^american\\s*$", "$options": "i"}},
     "grades": {"$elemMatch": {"score": {"$gt": 70}}},
    "address.coord.0": {"$lt": -65.754168}
}

proyeccion = {"name":1, "cuisine": 1, "grades":1, "address.coord":1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 7. Devolver los restaurantes que no preparan cocina americana y lograron un punto de calificación 'A' que no pertenece al distrito de Brooklyn. El documento debe mostrarse según la cocina en orden descendente.
filtro = {
    "cuisine": {"$not": {"$regex": "^american\\s*$", "$options": "i"}},
    "grades.grade": {"$eq": "A"},
    "borough": {"$not":{"$regex": "^brooklyn\\s*$", "$options": "i"}}
}

proyeccion = {"name":1, "cuisine": 1, "grades":1, "borough":1, "_id":0}
list(restaurantes.find(filtro, proyeccion).sort("cuisine", pymongo.DESCENDING))
```

```python
# 8. Devolver los restaurantes que pertenecen al distrito Bronx y preparan platos americanos o chinos
filtro = {
    "$or": [
        {"cuisine": {"$regex": "^american\\s*$", "$options": "i"}},
        {"cuisine": {"$regex": "^chinese\\s*$", "$options": "i"}}
    ],
    "borough": {"$regex": "^bronx\\s*$", "$options": "i"}
}


proyeccion = {"name":1, "cuisine": 1, "grades":1, "borough":1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
filtro = {
    "cuisine": {"$in": ["American ", "Chinese"] },
    "borough": {"$regex": "^bronx\\s*$", "$options": "i"}
}


proyeccion = {"name":1, "cuisine": 1, "grades":1, "borough":1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 9. Devolver ID del restaurante, nombre, distrito y la cocina para aquellos restaurantes que pertenecen al distrito de Staten Island o Queens o Bronx o Brooklyn
filtro = {
    "$or": [
        {"borough": {"$regex": "^staten island\\s*$", "$options": "i"}},
        {"borough": {"$regex": "^queens\\s*$", "$options": "i"}},
        {"borough": {"$regex": "^bronx\\s*$", "$options": "i"}},
        {"borough": {"$regex": "^queens\\s*$", "$options": "i"}}
    ]
}

proyeccion = {"restaurant_id": 1,"name":1, "borough":1, "cuisine":1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 10. Devolver ID del restaurante, nombre, distrito y la cocina de aquellos restaurantes que lograron una puntuación que no supere los 10
filtro = {
    "grades.score":{"$not": {"$gt": 10}}
}


proyeccion = {"restaurant_id": 1,"name":1, "borough":1, "cuisine":1,"grades.score":1, "_id":0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 11. Devolver ID del restaurante, el nombre y las calificaciones del restaurante para aquellos restaurantes que obtuvieron una calificación de "A" y obtuvieron un puntaje de 11 en una fecha ISO "2014-08-11T00: 00: 00Z" entre muchas fechas de encuesta
from datetime import datetime
import time

# Convertir la fecha ISO a un timestamp en milisegundos
iso_date = "2014-08-11T00:00:00Z"
fecha_objetivo = datetime.fromisoformat(iso_date.replace("Z", "+00:00"))
timestamp_ms = int(fecha_objetivo.timestamp() * 1000)


filtro = {'grades.grade': 'A',
          'grades.score': 11,
          'grades.date.$date': timestamp_ms}

proyeccion = {"restaurant_id": 1, "name": 1, "grades": 1, "_id": 0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 12. Devolver ID del restaurante, nombre, dirección y ubicación geográfica del restaurante de aquellos donde el segundo elemento de la matriz coord contiene un valor que es más de 42 y hasta 52
filtro = {
    "address.coord.1": {"$gt":42, "$lt": 52}
}


proyeccion = {"restaurant_id": 1, "name": 1, "address": 1, "_id": 0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 17. Busca los restaurantes cuyo nombre empiece por 'C'
filtro = {
    "name": {"$regex": "^c", "$options": "i"}
}


proyeccion = {"restaurant_id": 1, "name": 1, "address": 1, "_id": 0}
list(restaurantes.find(filtro, proyeccion))
```
