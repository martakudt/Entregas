```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
```

```python
driver = webdriver.Chrome()
ruta = "https://www.filmaffinity.com/es/ranking.php?rn=ranking_fa_movies"

driver.get(ruta)
```

```python
# Aceptar las cookies
boton_cookies = '//*[@id="qc-cmp2-ui"]/div[2]/div/button[3]'
boton_cookies = driver.find_element(By.XPATH, boton_cookies)
```

```python
boton_cookies.click()
```

```python
# Scroll
boton_scroll = '//*[@id="mt-content-cell"]/div[2]/div[3]'
boton_scroll = driver.find_element(By.XPATH, boton_scroll)
```

```python
# Scroll
for i in range(0,8):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1.5)
    boton_scroll.click()
    time.sleep(1.5)
```

```python
soup = bs(driver.page_source,"lxml")
soup
```

```python
titulo = []
ranking = []
anyo = []
director = []
puntuacion = []

for elemento in soup.select("div.mc-info-container"):
    titulo.append(elemento.select("div.mc-title")[0].text)
    anyo.append(elemento.select("span.mc-year")[0].text)
    director.append(elemento.select("div.mc-director")[0].text)

for elemento in soup.select("li.position"):
    ranking.append(elemento.text)

for elemento in soup.select("div.avg-rating"):
    puntuacion.append(elemento.text)

peliculas = pd.DataFrame({
    "titulo": titulo,
    "ranking": ranking,
    "año": anyo,
    "director": director,
    "puntuación": puntuacion
})

peliculas
```
